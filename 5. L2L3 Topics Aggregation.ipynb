```python
# Obtain L3 from L2 topics
def obtain_L3_pros(topic_num):
    '''
    topic num to be passed into this function has to match with those in the aggregation
    '''
    if topic_num in [-1]:
        return 'not included'
    if topic_num in [0, 1, 10, 22, 27]:
        return 'mixed topics'
    elif topic_num in [2, 18, 21, 26]:
        return 'benefits'
    elif topic_num in [3, 17]:
        return 'environment'
    elif topic_num in [4, 20]:
        return 'helpful colleagues'
    elif topic_num in [5,7,8,29]:
        return 'learning opportunities'
    elif topic_num in [6,9]:
        return 'good work life balance'
    elif topic_num in [11,13]:
        return 'brand'
    elif topic_num in [12]:
        return 'good place to work at'
    elif topic_num in [14]:
        return 'accomodates employees'
    elif topic_num in [15,16]:
        return 'good salary'
    elif topic_num in [19]:
        return 'technologies'
    elif topic_num in [23]:
        return 'nothing to share'
    elif topic_num in [24]:
        return 'many leaves'
    elif topic_num in [25]:
        return 'good culture'
    elif topic_num in [28]:
        return 'flexible working hours'
    
pros_topics['L2 Topic Name'] = pros_topics['L2 topic'].apply(obtain_L3_pros)
pros_topics
```

```python
def obtain_L3_cons(topic_num):
    if topic_num in [-1]:
        return 'not included'
    elif topic_num in [0, 1, 20, 26, 39, 48, 49]:
        return 'mixed topics'
    elif topic_num in [2, 13]:
        return 'long working hours'
    elif topic_num in [3, 10, 28, 32]:
        return 'high workload'
    elif topic_num in [4, 5, 21, 27, 37, 40]:
        return 'no cons that one can think of at the moment'
    elif topic_num in [6, 7, 14]:
        return 'low pay'
    elif topic_num in [8, 9, 29, 31, 45]:
        return 'no work life balance'
    elif topic_num in [11]:
        return 'no cons as of now'
    elif topic_num in [12,22,34,36]:
        return 'bad management'
    elif topic_num in [15,23]:
        return 'bad culture'
    elif topic_num in [16]:
        return 'culture'
    elif topic_num in [17]:
        return 'office problems'
    elif topic_num in [18]:
        return 'not a good place for those who want growth'
    elif topic_num in [19]:
        return 'difficulty getting converted to perm'
    elif topic_num in [24]:
        return 'slow career progression'
    elif topic_num in [25,30,33,41,46,47]:
        return 'nil'
    elif topic_num in [35]:
        return 'office politics'
    elif topic_num in [38]:
        return 'slow systems'
    elif topic_num in [42]:
        return 'little opportunities for contract staff'
    elif topic_num in [43]:
        return 'tight deadlines and stressful'
    elif topic_num in [44]:
        return 'high targets and high attrition rate'
cons_topics['L2 Topic Name'] = cons_topics['L2 topic'].apply(obtain_L3_cons)
cons_topics
```

### Running BERTopic to obtain L3 topics for pros

```python
def generate_L3_from_chosen_L2(L2_topic):
    '''
    Obtain L3 topic from L2 L1 topics excel file
    '''
    for topic in L2_topic:
        topic_of_interest = pros_topics[pros_topics['L2 Topic Name'] == topic]
        topic_of_interest_lis = topic_of_interest['document'].tolist()
        print(topic_of_interest_lis)
        roberta = TransformerDocumentEmbeddings('roberta-base')

        topic_model_pros = BERTopic(embedding_model=roberta, n_gram_range=(1,3), nr_topics=5, verbose=True, calculate_probabilities=True)
        topics_pros, probs_pros = topic_model_pros.fit_transform(topic_of_interest_lis) # BERTopic accepts only list
        topic_info_pros = topic_model_pros.get_topic_info()
        pros_df = pd.DataFrame({'L3 Topic': topics_pros, 'L2 Topic': topic,'document': topic_of_interest_lis, 'serial_number': topic_of_interest['serial_number']})
        pros_reviews_dir = r'C:\Users\jingh\OneDrive\Desktop\DBS Internship stuff\Employee Reviews Documents\Employee Reviews\DBS Reviews\All DBS Reviews\L3 Topics\DBS_L3_pros_reviews.xlsx'

        if os.path.isfile(pros_reviews_dir):
            full_df = pd.read_excel(pros_reviews_dir)
            full_df = pd.concat([full_df, pros_df], axis=0)
            full_df.to_excel(pros_reviews_dir, index=False)
        else:
            pros_df.to_excel(pros_reviews_dir, index=False)
            
        print(topic_model_pros.get_topic_info())
        topic_num = len(topic_model_pros.get_topic_info())
        keywords_pros = pd.DataFrame(columns=['L2 Topic', 'L3 Topic','Keyword1','Keyword2','Keyword3','Keyword4','Keyword5','Keyword6','Keyword7','Keyword8','Keyword9','Keyword10'])
        for i in range(topic_num):
            if -1 in topic_info_pros['Topic'].tolist():
                keywords_pros = keywords_pros.append({'L2 Topic': topic,
                                                     'L3 Topic':(i-1),
                                                     'Keyword1':topic_model_pros.get_topic(i-1)[0][0],
                                                     'Keyword2':topic_model_pros.get_topic(i-1)[1][0],
                                                     'Keyword3':topic_model_pros.get_topic(i-1)[2][0],
                                                     'Keyword4':topic_model_pros.get_topic(i-1)[3][0],
                                                     'Keyword5':topic_model_pros.get_topic(i-1)[4][0],
                                                     'Keyword6':topic_model_pros.get_topic(i-1)[5][0],
                                                     'Keyword7':topic_model_pros.get_topic(i-1)[6][0],
                                                     'Keyword8':topic_model_pros.get_topic(i-1)[7][0],
                                                     'Keyword9':topic_model_pros.get_topic(i-1)[8][0],
                                                     'Keyword10':topic_model_pros.get_topic(i-1)[9][0]},
                                                    ignore_index=True)
            else:
                keywords_pros = keywords_pros.append({'L2 Topic': topic,
                                                     'L3 Topic':(i),
                                                     'Keyword1':topic_model_pros.get_topic(i)[0][0],
                                                     'Keyword2':topic_model_pros.get_topic(i)[1][0],
                                                     'Keyword3':topic_model_pros.get_topic(i)[2][0],
                                                     'Keyword4':topic_model_pros.get_topic(i)[3][0],
                                                     'Keyword5':topic_model_pros.get_topic(i)[4][0],
                                                     'Keyword6':topic_model_pros.get_topic(i)[5][0],
                                                     'Keyword7':topic_model_pros.get_topic(i)[6][0],
                                                     'Keyword8':topic_model_pros.get_topic(i)[7][0],
                                                     'Keyword9':topic_model_pros.get_topic(i)[8][0],
                                                     'Keyword10':topic_model_pros.get_topic(i)[9][0]},
                                                    ignore_index=True)

        keywords_dir = r'C:\Users\jingh\OneDrive\Desktop\DBS Internship stuff\Employee Reviews Documents\Employee Reviews\DBS Reviews\All DBS Reviews\L3 Topics\DBS_L3_pros_all_keywords.xlsx'    
        if os.path.isfile(keywords_dir):
            existing_keywords = pd.read_excel(keywords_dir)
            all_keywords = pd.concat([existing_keywords, keywords_pros],axis=0)
            all_keywords.to_excel(keywords_dir, index=False)
        else:
            keywords_pros.to_excel(keywords_dir, index=False)

    return pros_df
pros_df = generate_L3_from_chosen_L2(['good work life balance', 'flexible working hours', 'brand', 'technologies', 'nothing to share', 'good culture'])
pros_df
```
