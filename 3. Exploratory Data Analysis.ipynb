### Descriptive analysis:
- How many positive and negative reviews 
- How many reviews for each year

```python
import pandas as pd
import matplotlib.pyplot as plt
from datetime import date
df = pd.read_csv('../Employee Reviews/DBS Reviews/All DBS Reviews/All_DBS_Reviews.csv')
df = df.drop_duplicates()
df = df[~pd.isna(df['Date'])]
df
```

```python
df.info()
```

### Number Of Reviews Per Year

```python
good_ratings = df[df['Rating'] >= 4]
neutral_ratings = df[df['Rating'] == 3]
subpar_ratings = df[df['Rating'] <= 2]
good_ratings
```

```python
neutral_ratings
```

```python
def get_year(dt_object):
    return dt_object[-5:-1]
good_ratings['Year'] = good_ratings['Date'].apply(get_year)
print(good_ratings['Year'].value_counts())
good_ratings['Year'].value_counts().plot(kind='bar')
```

```python
neutral_ratings
```

```python
neutral_ratings['Year'] = neutral_ratings['Date'].apply(get_year)
print(neutral_ratings['Year'].value_counts())
neutral_ratings['Year'].value_counts().plot(kind='bar')
```

```python
subpar_ratings['Year'] = subpar_ratings['Date'].apply(get_year)
print(subpar_ratings['Year'].value_counts())
subpar_ratings['Year'].value_counts().plot(kind='bar')
```
